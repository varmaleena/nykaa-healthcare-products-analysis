### find outliers in original prices

```python
Q1 = project["Original Price"].quantile(0.25)
Q3 = project["Original Price"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = project[(project["Original Price"] < lower_bound) | (project["Original Price"] > upper_bound)]
outliers
```

#### How many Nykaa healthcare products fall with in each range

```python
PriceRange = pd.cut(project["Original Price"], bins=[0,1000,2000,3000,4000,5000,10000,20000], right=False)
price_counts = PriceRange.value_counts().sort_index()
sns.barplot(price_counts)
plt.xticks(rotation=45)
plt.title("Product Count by Original Price Range")
```

### Which brand gives the biggest discounts on high-priced items?

```python
item=project[project["Original Price"]>10000]
items=item.groupby("Brands")["Discount in %"].max().sort_values(ascending=False)
items
```

```python
sns.barplot(x=items.values, y=items.index,orient="y")
plt.xlabel("Average Discount % on High-Priced Items")
plt.ylabel("Brands")
plt.title("Which Brand Gives Biggest Discounts on High-Priced Items?")
plt.show()
```
